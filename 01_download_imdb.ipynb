# 01_download_imdb

Download the IMDB dataset from Hugging Face `datasets` and save as CSV(s).

```python
# pip install datasets
from datasets import load_dataset
import os
out_dir = 'data/imdb'
os.makedirs(out_dir, exist_ok=True)
ds = load_dataset('imdb')
ds['train'].to_csv(os.path.join(out_dir, 'train.csv'), index=False)
ds['test'].to_csv(os.path.join(out_dir, 'test.csv'), index=False)
print('Saved to', out_dir)
```
