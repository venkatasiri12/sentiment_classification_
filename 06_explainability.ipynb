# 06_explainability

SHAP/LIME examples for TF-IDF (and notes for Transformer).

```
import os, joblib
art='artifacts/tfidf/lr_tfidf.joblib'
if os.path.exists(art):
    pkg = joblib.load(art)
    model = pkg['model']; vect = pkg['vectorizer']
    import shap
    sample = ['I loved the movie', 'Terrible and boring']
    X = vect.transform(sample)
    explainer = shap.LinearExplainer(model, X)
    sv = explainer.shap_values(X)
    print('Computed SHAP demo (small).')
else:
    print('Run TF-IDF notebook first.')
```
