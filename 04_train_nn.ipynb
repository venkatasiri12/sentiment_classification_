# 04_train_nn

Train a simple Embedding + LSTM and a small CNN using Keras.

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL']='2'
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Conv1D, GlobalMaxPooling1D, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
import pickle

DATA_DIR='data/imdb'
out='artifacts/nn'
os.makedirs(out, exist_ok=True)
train_df = pd.read_csv(os.path.join(DATA_DIR,'train.csv'))
tokenizer = Tokenizer(num_words=30000, oov_token='<OOV>')
tokenizer.fit_on_texts(train_df['text'])
maxlen=256
X = pad_sequences(tokenizer.texts_to_sequences(train_df['text']), maxlen=maxlen)
y = train_df['label'].values
model = Sequential([Embedding(30000,128,input_length=maxlen), LSTM(128), Dropout(0.3), Dense(1,activation='sigmoid')])
model.compile('adam','binary_crossentropy',['accuracy'])
model.fit(X, y, validation_split=0.1, epochs=1, batch_size=128)
model.save(os.path.join(out,'lstm.h5'))
with open(os.path.join(out,'tokenizer.pkl'),'wb') as f:
    pickle.dump(tokenizer,f)
print('Saved NN artifacts to', out)
```
